### SC-Track: Demo dataset running

To run our demo dataset, please download our demo image and annotation files from [here](https://zenodo.org/record/8310572). After unzip your download file, 
the dataset structure should be like this：

![image](images/demo_dataset_struct.png):


In this dataset, the `image.tif` is the raw image file, `mask.tif` and `annotation.json` both are segmented results for the `image.tif`, 

SC-Track can using this two annotation format to do the tracking. 

The two tracking_output folders are the running output `using annotation.json` and the `using mask.tif`. When you try to run this demo 

dataset yourself, a tracking_output folder will be generated in the same directory as `image.tif`. Its content should be the same as 

these two, and which one is the same depends on the mask segmentation annotation file you use, or json segmentation annotation file.


Now we assume that you unzip the demo dataset in `G:\raw data\demo`， and solved the environmental dependency problem according to [SC-Track's installation guidelines](./quick-start.ipynb/#Installation). 

You can using SC-Track to tracking this demo dataset, here is the using guidance: 

----------------
#### Using mask to do the tracking

````python
import tifffile

image = tifffile.imread(r"G:\raw data\demo\image.tif")

mask = tifffile.imread(r"G:\raw data\demo\mask.tif")

print(image.shape)
print(mask.shape)
````

````python
import matplotlib.pyplot as plt

plt.imshow(image[0], cmap='gray')
plt.axis('off')
````

````python
plt.imshow(mask[0], cmap='gray')
plt.axis('off')
````

**Each instance in the mask needs to ensure that the pixel value is unique**

````python
import numpy as np

np.unique(mask[0])
````

--------------
Now you can run this command to try SC-Track:

``` sctrack -i "G:\raw data\demo\image.tif" -a "G:\raw data\demo\mask.tif" ```

````python
# Please do not run this command directly unless you are using it directly in jupyter and have switched to the correct environment. 
# The recommended way is to copy this command to the terminal and run it (remember to remove `!`)

!sctrack -i "G:\raw data\demo\image.tif" -a "G:\raw data\demo\mask.tif"
````

The process will like this:
![output](images/sctrack_run_mask_process.png)

**Normally, when the visualization results are exported, running our demo data set will take no more than `5` minutes from start to finish; if the visualization results are not exported, the running time will not exceed `3` minutes.**

#### Using annotation.json to do the tracking

You also can using the json format annotation to do the tracking, this json file can be load by **VGG Image Annotator**, you can downlad it from [here](https://www.robots.ox.ac.uk/~vgg/software/via/downloads/via-2.0.12.zip) , and then you can load the `annotation.json` file. The format please see [quick-start](./quick-start.ipynb/).

Load result like this: 
![image](images/via_load.png)

**And then select the image file as the background, to show the annotation. As shown below, you can select all the images in `png` folder.**



![image](images/via_select_file.png)

**After that, you can see the segmentation result:**

![image](images/via_add_image.png)

Then you can run the same command to do the tracking:

``` sctrack -i "G:\raw data\demo\image.tif" -a "G:\raw data\demo\annotation.json" ```

````python
!sctrack -i "G:\raw data\demo\image.tif" -a "G:\raw data\demo\annotation.json"
````

The process will like this:
![output](images/sctrack_run_json_process.png)

Using json format can directly provide position contour information without calculation, which is faster than mask.

Now you can get the `tracking_output` result, its struct like this:
    
|__tracking_output\\\
&emsp;      ├─TrackTree\\\
&emsp;      └─track_visualization.tif\\\
&emsp;      └─track.csv\
&emsp;      └─result_with_track.json


![image](images/tracking_output.png)

The `TrackTree` folder contains the detailed information of each TrackTree built during the tracking process, the `track_visualization.tif` folder contains the visual image information of the tracking(if you close the visualization option, this content will not be generated), and `track.csv` is a detailed table of the tracking results. For specific information about the track.csv, see [https://github.com/chan-labsite/SC-Track/blob/master/notebook/quick-start.ipynb](https://github.com/chan-labsite/SC-Track/blob/master/notebook/quick-start.ipynb). The content of `result_with_track.json` is consistent with the structure of annotation.json, but result_with_track.json contains track information and smoothed classification information (if exists).
